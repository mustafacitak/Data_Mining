```python
# Aşağıdaki program veri okunarak LOF algoritmasına bağlı olarak aykırı değerlere sahip olup olmadığını 
# analiz etmektedir.

data = pd.read_csv("/Users/mustafacitak/Desktop/Python/Data_Science/diabetes.csv")

# Sınıf değişkeni ile diğerlerinin ayrılması
X = data.iloc[:, 0:8]
y = data.iloc[:, 8]
```

```python
# Veri kümesi okunduktan sonra sınıf özniteliği dışında kalan öznitelikler esas alınarak aykırı 
#değer analizleri yapılır.
# LOF algoritması ile aykırı değerleri belirlemek için LocalOutlierFactor() fonksiyonu kullanılır. 
#En yakın komşu sayısını belirlemek üzere n_neighbors=20 parametresi tanımlandı.

X = X.iloc[:, 0:4]
clf = LocalOutlierFactor(n_neighbors=20)

# Aykırı gözlemler için -1, diğerleri için 1 etiketini görüntüler.
aykırı = clf.fit_predict(X)

# Yapılan işlemler sonucunda LOF algoritmasına göre aykın değer olarak öngörülen gözlemler 
#aşağıda görüldüğü biçimde işaretlenir.
# Burada -1 ile belirtilen gözlemler aykırı değerlere işaret etmektedir.

print(aykırı[1:1001])

# Bu bölümde makine öğrenmesinin geleneksel bir yöntemi kullanılarak aykırı değerler öngörüldü.
# Veri kümelerindeki aykın değerleri bulma veya bir başka deyişle anomali belirleme amacıyla derin
#öğrenmenin ilginç yöntemlerinden de söz etmek mümkündür.
```
